Goal is to obtaining the USA housing price driving factors, then map and graph that information. 
To have a starting point I will use the Wikipedia list of country in the USA.
This is the code to screen scrap that current list from Wikipedia.


#https://en.wikipedia.org/wiki/List_of_United_States_counties_and_county_equivalents
#https://en.wikipedia.org/wiki/County_(United_States)

Referance code used: https://medium.com/@Alexander_H/scraping-wikipedia-with-python-8000fc9c9e6c

```python
import wikipedia
import requests
from bs4 import BeautifulSoup
import pandas as pd
import fnmatch
import numpy as np

# first pull the HTML from the page that links to all of the pages with the links.
# in this case, this page gives the links list pages of county in the USA.

website_url = requests.get('https://en.wikipedia.org/wiki/List_of_United_States_counties_and_county_equivalents').text
states = pd.read_csv('C:/Users/Joe/wikipedia_state_list.csv', encoding='cp1252',low_memory=False)

soup = BeautifulSoup(website_url,'lxml')
#print(soup.prettify())

My_table = soup.find('table',{'class':'wikitable sortable'})
links = My_table.findAll('a')

County =[]
for links in links:
    County.append(links.get('title'))
#display(County)  

df = pd.DataFrame()

df['County'] = County
df = df.drop_duplicates(keep=False, inplace=False)

# remove lists of non County names
df = df[~df.County.str.contains(',?? Metropolitan Statistical Area' )]
df = df[~df.County.str.contains('Micropolitan Statistical Area' )]
df = df[~df.County.str.contains('Puerto Rico')]
df = df[~df.County.str.contains('Guam')]
df = df[~df.County.str.contains('American Samoa')]

# split dataframe based on comma into two columns
df = df['County'].str.split(',', expand=True)
df.columns = ['County', 'State']

# remove rows with no data, i.e. leave only "Country" and "State" complete rows
df = df.dropna() 

# add in "District of Columbia" as a new row
index_len = len(df.County)
df = df.append(pd.DataFrame({'County': 'District of Columbia',
                            'State': 'District of Columbia'}, index=['']))

df = df.sort_values(by=['State', 'County'])
df = df.reset_index(drop=True)

print(df)

# Writting the new date to a file 
df.to_csv("C:/Users/Joe/wikipedia_county_list.csv", index=False)
```
